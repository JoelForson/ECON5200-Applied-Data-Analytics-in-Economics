```python
import pandas as pd
import numpy as np
np.random.seed(2026)
```

```python
population_size = 100000
true_mean_income = 75000
population_income = np.random.lognormal(mean = np.log(true_mean_income), sigma = 0.5, size = population_size).round(2)

print(np.mean(population_income))
```

```python
from numpy.random import random_sample
sample_size = 500
random_sample = np.random.choice(population_income, size = sample_size)

print(np.mean(random_sample))
```

```python
biased_sample = population_income[population_income > 80000]
observed_biased_sample = np.random.choice(biased_sample, size = sample_size)
print(np.mean(observed_biased_sample))
```

# **Algortithim R**

```python
import random
stream = range(1,1001)
k = 5
reservoir = []
```

```python
for i, item in enumerate(stream):
  if i < k:
    reservoir.append(item)
  else:
    j = random.randint(0,i)
    if j < k:
      reservoir[j] = item

print(reservoir)
```

# **Biased Population Data**

```python
population = np.concatenate([np.zeros(6000),np.ones(4000)])
```

```python
phone_prob = np.where(population == 0,0.10,0.90)
has_phone = np.random.rand(len(population)) < phone_prob
sample_frame = population[has_phone]
```

```python
sample_frame.mean()
```

# **The Heckman Logic**

```python
weights = np.where(sample_frame == 0, 1/0.1, 1/0.99)
```

```python
weighted_mean = np.average(sample_frame, weights = weights)
print(weighted_mean)
```
